```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x)
def f(x):
    return 1 / (1 + 25 * x**2)

# Set parameters
N = 2  # Degree of the interpolating polynomial
x_dense = np.linspace(-1, 1, 500)  # Dense grid for plotting the true function and interpolation

# Compute equally spaced grid points x_i and corresponding f(x_i) values
x_nodes = np.linspace(-1, 1, N + 1)  # Equally spaced nodes
y_nodes = f(x_nodes)

# Output the interpolation points and their corresponding function values
print("Points of Interpolation (x_i) and their corresponding f(x_i):")
for xi, yi in zip(x_nodes, y_nodes):
    print(f"x = {xi:.4f}, f(x) = {yi:.4f}")

# Find the interpolating polynomial coefficients
pN_coefficients = np.polyfit(x_nodes, y_nodes, N)  # Degree N polynomial
pN_values = np.polyval(pN_coefficients, x_dense)  # Evaluate polynomial on dense grid

# Plot the results
plt.figure(figsize=(8, 6))
plt.plot(x_dense, f(x_dense), 'r-', linewidth=2, label='$f(x) = \\frac{1}{1 + 25x^2}$')
plt.plot(x_dense, pN_values, 'b--', linewidth=2, label=f'Interpolating polynomial $p_{{{N}}}(x)$')
plt.plot(x_nodes, y_nodes, 'ko', markersize=8, label='Points of Interpolation')

# Add labels, legend, and display plot
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title(f'Runge\'s Phenomenon with Interpolating Polynomial of Degree {N}')
plt.legend()
plt.grid(True)
plt.axis([-1, 1, -1, 1])
plt.show()


   
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Define the function f(x)
def f(x):
    return 1 / (1 + 25 * x**2)

# Parameters
N_values = [2, 4, 6, 8, 10]  # Number of data points (xi)
n_values = [1, 2, 3, 4]      # Degrees of polynomials (qn) less than N

# Output directory
output_dir = 'demmanue/'  # Change this to your desired directory
os.makedirs(output_dir, exist_ok=True)  # Ensure the directory exists

# Define the grid for the data points
x_dense = np.linspace(-1, 1, 500)  # Fine grid for plotting the true function
for N in N_values:  # Number of data points
    x_nodes = np.linspace(-1, 1, N + 1)  # N+1 equally spaced points on [-1, 1]
    y_nodes = f(x_nodes)  # f(x) values at each node

    # Loop over degrees n of the polynomial (less than N)
    for n in n_values:
        if n >= N:
            continue

        # Construct the Vandermonde matrix for degree n polynomial
        A = np.vander(x_nodes, n + 1, increasing=True)

        # Solve the least squares problem: c = (A^T A)^(-1) A^T y
        coefficients = np.linalg.lstsq(A, y_nodes, rcond=None)[0]

        # Generate the fitted polynomial qn(x) on the fine grid
        qn_values = np.polyval(coefficients[::-1], x_dense)  # Reverse coefficients for np.polyval

        # Plot the results
        plt.figure(figsize=(8, 6))
        plt.plot(x_dense, f(x_dense), 'r-', linewidth=2, label='$f(x) = \\frac{1}{1 + 25x^2}$')  # True function
        plt.plot(x_dense, qn_values, 'b--', linewidth=2, label=f'Least squares fit $q_{n}(x)$ (degree {n})')

        # Plot the data nodes for clarity
        plt.plot(x_nodes, y_nodes, 'ko', markersize=8, label='Data nodes')

        # Add labels, legend, and display plot
        plt.xlabel('$x$')
        plt.ylabel('$y$')
        plt.title(f'Least Squares Fit for $f(x)$ with {N+1} nodes and degree {n}')
        plt.legend()
        plt.grid(True)
        plt.axis([-1, 1, -1, 1])
        plt.show()

        # Save the figure as JPG
        output_path = os.path.join(output_dir, f'least_squares_fit_N{N}_n{n}.jpg')  # Save with unique name
        plt.savefig(output_path, format='jpg')
        print(f"Saved plot for N={N} and n={n} at {output_path}")
```
